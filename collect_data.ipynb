## 1. Validation

```python
RT_1_CHECKPOINTS = {
    "rt_1_x": "rt_1_x_tf_trained_for_002272480_step",
    "rt_1_400k": "rt_1_tf_trained_for_000400120",
    "rt_1_58k": "rt_1_tf_trained_for_000058240",
    "rt_1_1k": "rt_1_tf_trained_for_000001120",
}

def get_rt_1_checkpoint(name, ckpt_dir="./SimplerEnv/checkpoints"):
  assert name in RT_1_CHECKPOINTS, name
  ckpt_name = RT_1_CHECKPOINTS[name]
  ckpt_path = os.path.join(ckpt_dir, ckpt_name)
  if not os.path.exists(ckpt_path):
    if name == "rt_1_x":
      !gsutil -m cp -r gs://gdm-robotics-open-x-embodiment/open_x_embodiment_and_rt_x_oss/{ckpt_name}.zip {ckpt_dir}
      !unzip {ckpt_dir}/{ckpt_name}.zip -d {ckpt_dir}
    else:
      !gsutil -m cp -r gs://gdm-robotics-open-x-embodiment/open_x_embodiment_and_rt_x_oss/{ckpt_name} {ckpt_dir}
  return ckpt_path

def validate_rt1_x(task_name="google_robot_pick_coke_can", model_name="rt_1_x", num_episodes=5):
  if 'env' in locals():
    print("Closing existing env")
    env.close()
    del env
  env = simpler_env.make(task_name)
  sapien.render_config.rt_use_denoiser = False  # Colab fix

  ckpt_path = get_rt_1_checkpoint(model_name)
  model = RT1Inference(saved_model_path=ckpt_path, policy_setup="google_robot" if "google" in task_name else "widowx_bridge")

  successes = 0
  for episode in range(num_episodes):
    obs, reset_info = env.reset()
    instruction = env.get_language_instruction()
    model.reset(instruction)
    print(f"Episode {episode + 1}, Instruction: {instruction}")

    image = get_image_from_maniskill2_obs_dict(env, obs)
    predicted_terminated, success, truncated = False, False, False
    timestep = 0
    while not (predicted_terminated or truncated):
      raw_action, action = model.step(image)
      predicted_terminated = bool(action["terminate_episode"][0] > 0)
      obs, reward, success, truncated, info = env.step(
          np.concatenate([action["world_vector"], action["rot_axangle"], action["gripper"]])
      )
      print(f"  Timestep {timestep}: {info}")
      image = get_image_from_maniskill2_obs_dict(env, obs)
      timestep += 1

    if success:
      successes += 1
    print(f"  Episode {episode + 1} success: {success}")

  print(f"Validation success rate: {successes / num_episodes}")
  env.close()
```

```python
validate_rt1_x(task_name="google_robot_pick_coke_can", model_name="rt_1_x", num_episodes=5)
```

## 2. Trajectory collection

```python
def collect_trajectories(
    task_name="google_robot_pick_coke_can",
    model_name="rt_1_x",
    num_episodes=100,
    output_dir="trajectories"
):
    if 'env' in locals():
        print("Closing existing env")
        env.close()
        del env
    env = simpler_env.make(task_name)
    sapien.render_config.rt_use_denoiser = False

    ckpt_path = get_rt_1_checkpoint(model_name)
    model = RT1Inference(saved_model_path=ckpt_path, policy_setup="google_robot" if "google" in task_name else "widowx_bridge")

    os.makedirs(output_dir, exist_ok=True)

    for episode in range(num_episodes):
        obs, reset_info = env.reset()
        instruction = env.get_language_instruction()
        model.reset(instruction)
        print(f"Collecting Episode {episode + 1}, Instruction: {instruction}")

        trajectory = {
            "images": [],
            "actions": [],
            "instructions": instruction,
        }

        image = get_image_from_maniskill2_obs_dict(env, obs)
        predicted_terminated, success, truncated = False, False, False
        timestep = 0
        while not (predicted_terminated or truncated):
            raw_action, action = model.step(image)
            predicted_terminated = bool(action["terminate_episode"][0] > 0)

            env_action = np.concatenate([action["world_vector"], action["rot_axangle"], action["gripper"]])

            obs, reward, success, truncated, info = env.step(env_action)

            trajectory["images"].append(image)
            trajectory["actions"].append(env_action)

            print(f"  Timestep {timestep}: {info}")

            image = get_image_from_maniskill2_obs_dict(env, obs)
            timestep += 1

        print(f"  Episode {episode + 1} finished. Success: {success}")

        filename = f"{task_name}_rt1_x_trajectory_{episode + 1:04d}.pkl" # Add zero-padding for easier sorting
        output_path = os.path.join(output_dir, filename)
        with open(output_path, "wb") as f:
            pickle.dump(trajectory, f)
        print(f"  Saved trajectory to {output_path}")

    print(f"Finished collecting and saving {num_episodes} trajectories to {output_dir}")
    env.close()

collect_trajectories(
    task_name="google_robot_pick_coke_can",
    model_name="rt_1_x",
    num_episodes=30,
    output_dir="trajectories",
)
```

## 2.5 Download trajectories

```python
import shutil
import os

def zip_directory(directory_path, output_zip_name):
    try:
        shutil.make_archive(output_zip_name.replace('.zip', ''), 'zip', directory_path)
        print(f"Successfully created '{output_zip_name}' from '{directory_path}'")
    except Exception as e:
        print(f"Error zipping directory '{directory_path}': {e}")

trajectories_dir = "trajectories"

import datetime
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_zip_file = f"robot_trajectories_{timestamp}.zip"

if os.path.exists(trajectories_dir) and os.path.isdir(trajectories_dir):
    zip_directory(trajectories_dir, output_zip_file)
else:
    print(f"Directory '{trajectories_dir}' does not exist. Please collect trajectories first.")

from google.colab import files
files.download(output_zip_file)
```
